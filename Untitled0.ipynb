```python
# Step 1: Check what's inside the 'sex' column
print("Unique values in 'sex' column:", data['sex'].unique())

# OPTIONAL: If you're not sure, stop here and check the output.
# Pick a value that represents the total/both sexes (like 'Total', 'All', 'Both', etc.)

# Step 2: Choose appropriate sex filter
# Update this line after checking your actual values ↓↓↓
filtered_sex_value = 'Total'  # <-- Change this based on what you see from Step 1

# Step 3: Prepare clean data
selected_countries = ['India', 'Nigeria', 'Bangladesh', 'Kenya']

line_data = data[
    (data['country_x'].isin(selected_countries)) &
    (data['sex'] == filtered_sex_value)
].copy()

# Step 4: Average duplicate rows for same country & year
line_data = line_data.groupby(['country_x', 'time_period'], as_index=False)['obs_value'].mean()

# Step 5: Plot Line Chart
fig_line = px.line(
    line_data,
    x='time_period',
    y='obs_value',
    color='country_x',
    markers=True,
    labels={
        "time_period": "Year",
        "obs_value": "Diarrhoea treatment (%)",
        "country_x": "Country"
    },
    title="Trends in Diarrhoea Treatment Coverage (2000–2022)"
)

fig_line.update_layout(
    hoverlabel=dict(
        bgcolor="white",
        font_size=14,
        font_family="Arial"
    )
)

# Show chart
fig_line.show()

# Save PNG
fig_line.write_image("diarrhoea_treatment_trends.png")
```

```python
# Step 1: Check unique 'sex' values again (just in case)
print("Unique values in 'sex' column:", data['sex'].unique())

# Step 2: Set the correct sex filter (same as before)
filtered_sex_value = 'Total'  # Change if needed

# Step 3: Filter data for year 2022 and selected sex
bar_data = data[
    (data['time_period'] == 2022) &
    (data['sex'] == filtered_sex_value)
].copy()

# Step 4: Group and average if duplicates exist
bar_data_grouped = bar_data.groupby('country_x', as_index=False)['obs_value'].mean()

# Step 5: Get Top 10 countries
top10 = bar_data_grouped.sort_values(by='obs_value', ascending=False).head(10)

# Step 6: Plot Bar Chart
fig_bar = px.bar(
    top10,
    x='country_x',
    y='obs_value',
    color='country_x',
    text='obs_value',
    labels={
        "country_x": "Country",
        "obs_value": "Diarrhoea treatment (%)"
    },
    title="Top 10 Countries by Diarrhoea Treatment Coverage (2022)"
)

fig_bar.update_traces(texttemplate='%{text:.1f}%', textposition='outside')

# 👇 Hide the legend
fig_bar.update_layout(
    showlegend=False,
    uniformtext_minsize=8,
    uniformtext_mode='hide',
    yaxis=dict(title="Diarrhoea treatment (%)"),
    xaxis=dict(title="Country"),
    hoverlabel=dict(
        bgcolor="white",
        font_size=14,
        font_family="Arial"
    )
)

# Show Bar Chart
fig_bar.show()

# Save the Bar Chart
fig_bar.write_image("top10_diarrhoea_treatment_2022.png")
```

```python
# Step 1: Set year and sex
filtered_year = 2022
filtered_sex = 'Total'  # Update if needed

# Step 2: Filter data
scatter_data = data[
    (data['time_period'] == filtered_year) &
    (data['sex'] == filtered_sex)
].copy()

# Step 3: Drop missing GDP or treatment values
scatter_data = scatter_data.dropna(subset=["GDP per capita (constant 2015 US$)", "obs_value"])

# Step 4: Group by country only (average values)
scatter_data = scatter_data.groupby('country_x', as_index=False).agg({
    "GDP per capita (constant 2015 US$)": "mean",
    "obs_value": "mean"
})

# Step 5: Create scatter plot
fig_scatter = px.scatter(
    scatter_data,
    x="GDP per capita (constant 2015 US$)",
    y="obs_value",
    text="country_x",
    trendline="ols",
    labels={
        "GDP per capita (constant 2015 US$)": "GDP per capita (US$)",
        "obs_value": "Diarrhoea treatment (%)"
    },
    title="Relationship Between GDP per Capita and Diarrhoea Treatment (2022)"
)

fig_scatter.update_traces(marker=dict(size=8, color='blue'), textposition='top center')

fig_scatter.update_layout(
    hoverlabel=dict(
        bgcolor="white",
        font_size=14,
        font_family="Arial"
    )
)

# Show plot
fig_scatter.show()

# Save as PNG
fig_scatter.write_image("gdp_vs_diarrhoea_treatment_2022.png")
```
